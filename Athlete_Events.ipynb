statistical information for numerical data in athlets dataset:

if we look at the count for each numerical feature we notice that ID have no missing value, Age have 9474 missing value,

if we look at the mean for each numerical feature we notice the the Average Age is 25, Average Height is 175, and Average Weight 70.

if we look at the std (standard deviation) --> we notice that each of Age, Height, and Weight are clustered around the mean because there value are low.

if we look at min (minimum) --> we notice that minimum value for Age is 10 years, minimum value for Height is 127 cm,
minimum value for Weight is 25 kilo, and the athlet data start since 1896.

if we look at 25% (lower quartile) --> we notice that 25% of Age are lie below 21, 25% of height are lie below 168, and 25% of Weight are lie below 60.

if we look at 50% (Median) --> we notice that half of the Age lie below 24 year and anothe half lie above 24 year, half of Height lie below 175 cm and other half lie above 175 cm, and half of weight lie below 70 kilo and other half lie above 70 kilo.

if we look at 75% (upper quartile) --> we notice that 25% of age lie above 28 year, 25% Height lie above 183 cm, and 25% Weight lie above 79 kilo.

if we look at max (maximum) --> we notice that maximum age is 97 years, maximum Height 226 cm, maximum Weight 214 kilo, and the athlets data ends in 2016.
